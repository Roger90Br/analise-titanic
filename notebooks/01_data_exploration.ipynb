```python
import seaborn as sns
import matplotlib.pyplot as plt

#Distribuição de Idade
sns.histplot(train_data['Age'].dropna(), kde=True)
plt.title('Distribuição de Idade')
plt.show()
```

```python
# Distribuição de Sobriviventes
sns.countplot(x='Survived', data=train_data)
plt.title('Distribuição de Sobreviventes')
plt.show()
```

```python
# Sobreviventes por Classe
sns.countplot(x='Pclass', hue='Survived', data=train_data)
plt.title('Sobreviventes por Classe')
plt.show()
```

```python
# Sobreviventes por Sexo
sns.countplot(x='Sex', hue='Survived', data=train_data)
plt.title('Sobreviventes por Porto de Sexo')
```

```python
# Sobreviventes por Porto de Embarque
sns.countplot(x='Embarked', hue='Survived', data=train_data)
plt.title('Sobreviventes por Porto de Embarque')
plt.show()
```

```python
# Boxplot de Idade por Classe e Sobrevivência
sns.boxplot(x='Pclass', y='Age', hue='Survived', data=train_data)
plt.title('Distribuição de Tarifa por Classe e Sobrevivência')
plt.show()
```

```python
# Pairplot entre variáveis numéricas
sns.pairplot(train_data[['Age', 'Fare', 'Pclass', 'Survived']].dropna(), hue='Survived')
plt.show()
```

```python
# Boxplot de Fare para identificar outliers
sns.boxplot(x='Fare', data=train_data)
plt.title('Boxplot de Fare')
plt.show()
```
